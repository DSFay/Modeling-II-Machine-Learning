###### Import the food production dataset and manipulate the dataset

```python
# Fay had to use an extra argument to re-encode both data sets from the Food and Agriculture Organization - re "latin-1"

df_production = pd.read_csv('../data/Production_Crops_Livestock_E_All_Data_NOFLAG.csv',encoding='latin-1')
```

```python
df_production
```

```python
# Clean redundant data
df_production= df_production.drop(['Area Code','Item Code','Item','Element Code'],axis = 1)
```

```python
# drop na
df_production.dropna(axis = 0,inplace=True)
```

```python
df_production.Element.unique()
```

```python
# Unpivot a DataFrame from wide to long format
df_production_melt = pd.melt(df_production,id_vars=['Area','Element','Unit'])
df_production_melt.head()
```

```python
df_production_melt.dropna(axis = 0, inplace=True)
```

```python
# transfer the variables in one column to seperate columns
df_production_melt_group = df_production_melt.groupby(['Element','Area','variable']).mean().reset_index()
```

```python
df_production_melt_group
```

```python
# change the Y-year to year

a = []
for i in range(len(df_production_melt_group)):

    year = df_production_melt_group.variable[i][1:]
    print(year)
    a.append(year)
a
```

```python
df_production_melt_group['year']=a
df_production_melt_group.tail()
```

```python
df_production_melt_group.drop(['variable'],axis = 1,inplace= True)
```

```python
df_production_melt_group.head()
```

```python
df_production_melt_group.Element.value_counts()
```

```python
df_production_melt_group.info()
```

```python
# We pick six productions Production, Yield, and Area Harvested

# 'Area harvested', 'Yield', 'Production', 'Stocks', 'Laying','Producing Animals/Slaughtered', 
    #'Yield/Carcass Weight', 'Milk Animals']

df_production_melt_group_six = df_production_melt_group[df_production_melt_group.Element.isin(['Yield','Production','Area harvested','Stocks','Milk Animals','Producing Animals/Slaughtered'])]
```

```python
df_production_melt_group_six
```

```python
# pivoting the df so that each of our three selected 'Element' types has it's own column populated with
# the corresponding agricultural production 'value'

df_production_melt_group_six_pv= pd.pivot_table(data=df_production_melt_group_six,index = ['Area','year'],values = ['value'],columns='Element' ).reset_index()
```

```python
df_production_melt_group_six_pv.head()
```

```python
# getting rid of the 'value' column as the first level of hierarchial column structure to simplify the df

df_production_melt_group_six_pv.columns = df_production_melt_group_six_pv.columns.droplevel()
```

```python
df_production_melt_group_six_pv.columns
```

```python
df_production_melt_group_six_pv
```

```python
# renaming the columns

df_production_melt_group_six_pv.columns = ['Country', 'Year', 'Area harvested', 'Production', 'Yield','Stocks','Milk Animals','Meat Animals']
```

```python
# taking a look at the result

df_production_melt_group_six_pv
```

```python
# checking the data types - year is an object

df_production_melt_group_six_pv.info()
```

```python
# telling py to read 'Year' as a numeric field

df_production_melt_group_six_pv['Year']=pd.to_numeric(df_production_melt_group_six_pv['Year'],errors='coerce')
```

###### import the nino dataset and manipulate the data

```python
nino = pd.read_csv('../data/detrend_nino34_monthly averages_monthly anomalies_1950-2022.csv')
nino.head()
```

```python
# the 2nd column of the 'nino' df has 5 columns contained inside of it

nino.columns
```

```python
# separating out the 5 separate columns from column 2 of the original nino df and renaming them

nino[['Year', 'Month', 'Total','Climate Adjusted','Anomaly']] = nino[' YR   MON  TOTAL ClimAdjust ANOM '].str.split(' +', 4, expand=True)
nino
```

```python
# dropping the original smooshed data and saving the tidy version

nino_tidy = nino.drop(['Unnamed: 0', ' YR   MON  TOTAL ClimAdjust ANOM '],axis=1)
nino_tidy
```

```python
nino_tidy.info()
```

```python
# converting our features from objects to numeric data types

nino_tidy['Year']=pd.to_numeric(nino_tidy['Year'],errors='coerce')
nino_tidy['Month']=pd.to_numeric(nino_tidy['Month'],errors='coerce')
nino_tidy['Total']=pd.to_numeric(nino_tidy['Total'],errors='coerce')
nino_tidy['Climate Adjusted']=pd.to_numeric(nino_tidy['Climate Adjusted'],errors='coerce')
nino_tidy['Anomaly']=pd.to_numeric(nino_tidy['Anomaly'],errors='coerce')
```

```python
nino_tidy.info()
```

```python
# The nino data has Sea Surface Temperatures (SST's) aggregated by month
# reaggregating at the yearly level to match our other data
   
        # The SST anomaly value indicates whether a neutral year, an El Nino year, or a La Nina year is present
            # I.E. the ENSO state - the state of the El Nino Southern Oscillation

nino_year = nino_tidy.copy()
```

```python
# taking the mean of all months to create a yearly average SST anomaly

nino_year = nino_year.groupby('Year').Anomaly.mean().reset_index()
nino_year.head()
```

```python
# checking for na's

nino_tidy.isna().sum()
```

```python
# Merge the fourth dataset

df_merge_4 = pd.merge(df_merge_3,nino_year,on='Year',how='left')
df_merge_4
```
